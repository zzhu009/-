```python
#描述性统计
import pandas as pd
import matplotlib.pyplot as plt

# 1. 读取数据（csv格式）
df1 = pd.read_csv('amazon_eco-friendly_products_final.csv')
df2 = pd.read_csv('amazon_eco_products_数据爬取原始生成_final.csv')
df3 = pd.read_csv('Form-responses-on-sustainability-products_final.csv')

# 2. 亚马逊绿色产品主数据集（df1）
print("主数据集-数值型变量描述性统计：")
print(df1[['price', 'rating', 'reviews']].describe())

print("\n主数据集-产品类别分布：")
print(df1['category'].value_counts())

print("\n主数据集-品牌分布：")
print(df1['brand'].value_counts().head(10))

print("\n主数据集-材质分布：")
print(df1['material'].value_counts().head(10))

# 3. 亚马逊绿色产品爬取数据集（df2）
print("\n爬取数据集-数值型变量描述性统计：")
print(df2[['price', 'rating', 'reviews']].describe())

feature_keywords = ['eco-friendly', 'sustainable', 'biodegradable', 'recycled', 'green']
for kw in feature_keywords:
    df2[kw] = df2['features'].astype(str).str.contains(kw, case=False, na=False)
    print(f"{kw} 产品数量：{df2[kw].sum()}")

print("\n爬取数据集-环保认证分布：")
print(df2['eco_certifications'].value_counts())

# 4. 可持续产品消费者问卷调查数据集（df3）
print("\n问卷数据集-购买意愿分数描述性统计：")
print(df3['purchase_likelihood'].describe())

print("\n问卷数据集-环保意识分布：")
print(df3['env_consideration'].value_counts())

print("\n问卷数据集-溢价意愿分布：")
print(df3['pay_premium'].value_counts())

print("\n问卷数据集-回收行为分布：")
print(df3['recycle_packaging'].value_counts())

print("\n问卷数据集-购买障碍统计（前10项）：")
print(df3['purchase_barriers'].value_counts().head(10))

# 5. 生成关键可视化
plt.hist(df1['price'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('绿色产品价格分布')
plt.xlabel('价格')
plt.ylabel('产品数量')
plt.show()

plt.boxplot(df3['purchase_likelihood'].dropna())
plt.title('消费者绿色产品购买意愿箱线图')
plt.ylabel('购买意愿（1-10分）')
plt.show()
```

```python
#相关性分析
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号

# 1. 读取数据
df1 = pd.read_csv('amazon_eco-friendly_products_final.csv')
df2 = pd.read_csv('amazon_eco_products_数据爬取原始生成_final.csv')
df3 = pd.read_csv('Form-responses-on-sustainability-products_final.csv')

# 2. 数据预处理与特征工程
# df2: 统计环保特性数量
if 'features' in df2.columns:
    df2['feature_count'] = df2['features'].astype(str).apply(lambda x: len(eval(x)) if x != 'nan' else 0)

# df2: 生成环保标签（如eco-friendly）哑变量
def has_eco_friendly(x):
    try:
        return int('eco-friendly' in [i.strip().lower() for i in eval(x)])
    except:
        return 0
if 'features' in df2.columns:
    df2['eco_friendly'] = df2['features'].apply(has_eco_friendly)

# df3: 将部分变量转为数值型
for col in ['purchase_likelihood', 'plastic_bottle_likelihood', 'cert_label_importance', 'willing_change_habit']:
    if col in df3.columns:
        df3[col] = pd.to_numeric(df3[col], errors='coerce')

# df3: 环保意识、溢价意愿、回收行为等转为0/1
for col in ['env_consideration', 'pay_premium', 'recycle_packaging', 'willing_change_habit']:
    if col in df3.columns:
        df3[col] = df3[col].map({'yes': 1, 'no': 0, 'sometimes': 0.5, 'maybe': 0.5, '未知': np.nan})

# 3. 相关性分析
def plot_corr(df, cols, title):
    corr = df[cols].corr()
    print(f"{title}相关性矩阵：")
    print(corr)
    plt.figure(figsize=(8,6))
    sns.heatmap(corr, annot=True, cmap='YlGnBu')
    plt.title(title + '相关性热力图')
    plt.show()

# 主数据集
cols1 = ['price', 'rating', 'reviews']
plot_corr(df1, [c for c in cols1 if c in df1.columns], '主数据集')

# 爬取数据集
cols2 = ['price', 'rating', 'reviews', 'feature_count', 'eco_friendly']
plot_corr(df2, [c for c in cols2 if c in df2.columns], '爬取数据集')

# 问卷数据集
cols3 = ['purchase_likelihood', 'env_consideration', 'pay_premium', 'recycle_packaging', 'cert_label_importance', 'willing_change_habit', 'plastic_bottle_likelihood']
plot_corr(df3, [c for c in cols3 if c in df3.columns], '问卷数据集')

# 如需对品类、品牌、材质等哑变量与因变量均值做分组对比，可用如下代码（以df1为例）：
if 'category' in df1.columns:
    print(df1.groupby('category')[['rating', 'reviews']].mean())
    df1.groupby('category')[['rating', 'reviews']].mean().plot(kind='bar')
    plt.title('不同品类下的平均评分和评论数')
    plt.ylabel('均值')
    plt.show()
```

```python
#基准回归
import pandas as pd
import statsmodels.api as sm

# 1. 读取数据
df1 = pd.read_csv('amazon_eco-friendly_products_final.csv')
df2 = pd.read_csv('amazon_eco_products_数据爬取原始生成_final.csv')
df3 = pd.read_csv('Form-responses-on-sustainability-products_final.csv')

# ========== 1. 主数据集：产品属性对市场表现的回归 ==========
df1 = df1.dropna(subset=['rating', 'price', 'reviews', 'category', 'material', 'brand'])
X1 = df1[['price', 'reviews']]
X1 = pd.concat([
    X1,
    pd.get_dummies(df1['category'], prefix='cat', drop_first=True, dtype=float),
    pd.get_dummies(df1['material'], prefix='mat', drop_first=True, dtype=float),
    pd.get_dummies(df1['brand'], prefix='brand', drop_first=True, dtype=float)
], axis=1)
X1 = sm.add_constant(X1)
y1 = pd.to_numeric(df1['rating'], errors='coerce')

# 检查数据类型
print("X1 dtypes:\n", X1.dtypes.value_counts())
print("y1 dtype:", y1.dtype)

# 去除所有含有缺失值的行
X1 = X1.loc[y1.notnull()]
y1 = y1.loc[y1.notnull()]

# 再次确保全部为float
X1 = X1.astype(float)
y1 = y1.astype(float)

model1 = sm.OLS(y1, X1).fit()
print("主数据集回归结果：")
print(model1.summary())

# ========== 2. 爬取数据集：产品特性对市场表现的回归 ==========
df2 = df2.dropna(subset=['rating', 'price', 'reviews', 'features'])
def feature_count(x):
    try:
        return len(eval(x)) if isinstance(eval(x), list) else len(str(x).split(','))
    except:
        return len(str(x).split(','))
df2['feature_count'] = df2['features'].astype(str).apply(feature_count)
keywords = ['eco-friendly', 'sustainable', 'biodegradable', 'recycled', 'certified', 'organic', 'green']
for kw in keywords:
    df2[kw] = df2['features'].astype(str).apply(lambda x: int(kw in x.lower()))
X2 = df2[['price', 'reviews', 'feature_count'] + keywords]
X2 = sm.add_constant(X2)
y2 = pd.to_numeric(df2['rating'], errors='coerce')

print("X2 dtypes:\n", X2.dtypes.value_counts())
print("y2 dtype:", y2.dtype)

X2 = X2.loc[y2.notnull()]
y2 = y2.loc[y2.notnull()]
X2 = X2.astype(float)
y2 = y2.astype(float)

model2 = sm.OLS(y2, X2).fit()
print("\n爬取数据集回归结果：")
print(model2.summary())

# ========== 3. 问卷数据集：消费者特征对购买意愿的回归 ==========
df3 = df3.dropna(subset=['purchase_likelihood', 'env_consideration', 'pay_premium', 'recycle_packaging', 'cert_label_importance', 'willing_change_habit'])
X3 = pd.get_dummies(df3[['env_consideration', 'pay_premium', 'recycle_packaging', 'cert_label_importance', 'willing_change_habit']], drop_first=True, dtype=float)
X3 = sm.add_constant(X3)
y3 = pd.to_numeric(df3['purchase_likelihood'], errors='coerce')

print("X3 dtypes:\n", X3.dtypes.value_counts())
print("y3 dtype:", y3.dtype)

X3 = X3.loc[y3.notnull()]
y3 = y3.loc[y3.notnull()]
X3 = X3.astype(float)
y3 = y3.astype(float)

model3 = sm.OLS(y3, X3).fit()
print("\n问卷数据集回归结果：")
print(model3.summary())
```

```python
#稳健性检验
import pandas as pd
import statsmodels.api as sm

# 读取数据
df1 = pd.read_csv('amazon_eco-friendly_products_final.csv')
df2 = pd.read_csv('amazon_eco_products_数据爬取原始生成_final.csv')
df3 = pd.read_csv('Form-responses-on-sustainability-products_final.csv')

# ========== 1. 主数据集：剔除极端值后的回归 ==========
# 剔除价格和评论数的上下1%极端值
df1_clean = df1[
    (df1['price'].between(df1['price'].quantile(0.01), df1['price'].quantile(0.99))) &
    (df1['reviews'].between(df1['reviews'].quantile(0.01), df1['reviews'].quantile(0.99)))
].dropna(subset=['rating', 'price', 'reviews', 'category', 'material', 'brand'])
X1 = df1_clean[['price', 'reviews']]
X1 = pd.concat([
    X1,
    pd.get_dummies(df1_clean['category'], prefix='cat', drop_first=True, dtype=float),
    pd.get_dummies(df1_clean['material'], prefix='mat', drop_first=True, dtype=float),
    pd.get_dummies(df1_clean['brand'], prefix='brand', drop_first=True, dtype=float)
], axis=1)
X1 = sm.add_constant(X1)
y1 = pd.to_numeric(df1_clean['rating'], errors='coerce')
X1 = X1.loc[y1.notnull()]
y1 = y1.loc[y1.notnull()]
X1 = X1.astype(float)
y1 = y1.astype(float)
model1_robust = sm.OLS(y1, X1).fit()
print("主数据集（剔除极端值）稳健性回归结果：")
print(model1_robust.summary())

# ========== 2. 爬取数据集：更换因变量为评论数的回归 ==========
df2 = df2.dropna(subset=['rating', 'price', 'reviews', 'features'])
def feature_count(x):
    try:
        return len(eval(x)) if isinstance(eval(x), list) else len(str(x).split(','))
    except:
        return len(str(x).split(','))
df2['feature_count'] = df2['features'].astype(str).apply(feature_count)
keywords = ['eco-friendly', 'sustainable', 'biodegradable', 'recycled', 'certified', 'organic', 'green']
for kw in keywords:
    df2[kw] = df2['features'].astype(str).apply(lambda x: int(kw in x.lower()))
X2 = df2[['price', 'feature_count'] + keywords]
X2 = sm.add_constant(X2)
y2 = pd.to_numeric(df2['reviews'], errors='coerce')  # 用评论数作为因变量
X2 = X2.loc[y2.notnull()]
y2 = y2.loc[y2.notnull()]
X2 = X2.astype(float)
y2 = y2.astype(float)
model2_robust = sm.OLS(y2, X2).fit()
print("\n爬取数据集（因变量换为评论数）稳健性回归结果：")
print(model2_robust.summary())

# ========== 3. 问卷数据集：将购买意愿分组为高低（分组回归） ==========
# 以购买意愿中位数为界，分为高意愿和低意愿两组
df3 = df3.dropna(subset=['purchase_likelihood', 'env_consideration', 'pay_premium', 'recycle_packaging', 'cert_label_importance', 'willing_change_habit'])
median_likelihood = df3['purchase_likelihood'].median()
df3['high_likelihood'] = (df3['purchase_likelihood'] > median_likelihood).astype(int)
X3 = pd.get_dummies(df3[['env_consideration', 'pay_premium', 'recycle_packaging', 'cert_label_importance', 'willing_change_habit']], drop_first=True, dtype=float)
X3 = sm.add_constant(X3)
y3 = df3['high_likelihood']  # 因变量为高意愿（1）/低意愿（0）
X3 = X3.loc[y3.notnull()]
y3 = y3.loc[y3.notnull()]
X3 = X3.astype(float)
y3 = y3.astype(float)
model3_robust = sm.Logit(y3, X3).fit()
print("\n问卷数据集（高低意愿分组Logit）稳健性回归结果：")
print(model3_robust.summary())
```

```python
#异质性分析
import pandas as pd
import statsmodels.api as sm

# 读取数据
df1 = pd.read_csv('amazon_eco-friendly_products_final.csv')
df2 = pd.read_csv('amazon_eco_products_数据爬取原始生成_final.csv')
df3 = pd.read_csv('Form-responses-on-sustainability-products_final.csv')

# ========== 1. 主数据集：按价格高低分组异质性回归 ==========
# 以价格中位数为界，分为高价组和低价组
median_price = df1['price'].median()
df1_high = df1[df1['price'] > median_price].dropna(subset=['rating', 'price', 'reviews', 'category', 'material', 'brand'])
df1_low = df1[df1['price'] <= median_price].dropna(subset=['rating', 'price', 'reviews', 'category', 'material', 'brand'])

def run_ols(df):
    X = df[['price', 'reviews']]
    X = pd.concat([
        X,
        pd.get_dummies(df['category'], prefix='cat', drop_first=True, dtype=float),
        pd.get_dummies(df['material'], prefix='mat', drop_first=True, dtype=float),
        pd.get_dummies(df['brand'], prefix='brand', drop_first=True, dtype=float)
    ], axis=1)
    X = sm.add_constant(X)
    y = pd.to_numeric(df['rating'], errors='coerce')
    X = X.loc[y.notnull()]
    y = y.loc[y.notnull()]
    X = X.astype(float)
    y = y.astype(float)
    model = sm.OLS(y, X).fit()
    return model

print("主数据集高价组异质性回归结果：")
print(run_ols(df1_high).summary())
print("\n主数据集低价组异质性回归结果：")
print(run_ols(df1_low).summary())

# ========== 2. 爬取数据集：按是否含有“eco-friendly”标签分组 ==========
df2 = df2.dropna(subset=['rating', 'price', 'reviews', 'features'])
df2['eco_friendly'] = df2['features'].astype(str).apply(lambda x: int('eco-friendly' in x.lower()))
df2_eco = df2[df2['eco_friendly'] == 1]
df2_non_eco = df2[df2['eco_friendly'] == 0]

def run_ols2(df):
    def feature_count(x):
        try:
            return len(eval(x)) if isinstance(eval(x), list) else len(str(x).split(','))
        except:
            return len(str(x).split(','))
    df['feature_count'] = df['features'].astype(str).apply(feature_count)
    keywords = ['eco-friendly', 'sustainable', 'biodegradable', 'recycled', 'certified', 'organic', 'green']
    for kw in keywords:
        df[kw] = df['features'].astype(str).apply(lambda x: int(kw in x.lower()))
    X = df[['price', 'reviews', 'feature_count'] + keywords]
    X = sm.add_constant(X)
    y = pd.to_numeric(df['rating'], errors='coerce')
    X = X.loc[y.notnull()]
    y = y.loc[y.notnull()]
    X = X.astype(float)
    y = y.astype(float)
    model = sm.OLS(y, X).fit()
    return model

print("\n爬取数据集含eco-friendly标签组异质性回归结果：")
print(run_ols2(df2_eco).summary())
print("\n爬取数据集不含eco-friendly标签组异质性回归结果：")
print(run_ols2(df2_non_eco).summary())

# ========== 3. 问卷数据集：按环保意识分组异质性回归 ==========
df3 = df3.dropna(subset=['purchase_likelihood', 'env_consideration', 'pay_premium', 'recycle_packaging', 'cert_label_importance', 'willing_change_habit'])
df3_yes = df3[df3['env_consideration'] == 'yes']
df3_no = df3[df3['env_consideration'] != 'yes']

def run_ols3(df):
    X = pd.get_dummies(df[['pay_premium', 'recycle_packaging', 'cert_label_importance', 'willing_change_habit']], drop_first=True, dtype=float)
    X = sm.add_constant(X)
    y = pd.to_numeric(df['purchase_likelihood'], errors='coerce')
    X = X.loc[y.notnull()]
    y = y.loc[y.notnull()]
    X = X.astype(float)
    y = y.astype(float)
    model = sm.OLS(y, X).fit()
    return model

print("\n问卷数据集环保意识为yes组异质性回归结果：")
print(run_ols3(df3_yes).summary())
print("\n问卷数据集环保意识非yes组异质性回归结果：")
print(run_ols3(df3_no).summary())
```
